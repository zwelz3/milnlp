```python
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
from sumy.summarizers.lex_rank import LexRankSummarizer as Summarizer  # use this summarizer for naive metadata summary
#
from milnlp.tokenizers import Tokenizer
from milnlp.collection.collection import Collection

LANGUAGE = "english"
stemmer = Stemmer(LANGUAGE)
summarizer = Summarizer(stemmer)
summarizer.stop_words = get_stop_words(LANGUAGE)
token = Tokenizer(LANGUAGE)
#
term_frequency_threshold = 0.5  
num_sentences = 8  # for composite summary
#
collectionpath = r"C:\Users\zwelz3\Documents\GTRI_Projects\ECCT_EW_EMS\Market Research"
```

```python
from milnlp.collection.collection import get_items

cobj = Collection(collectionpath)
cobj.flist, cobj.dlist = get_items(collectionpath, set(), {collectionpath: 0})
```

### Run and verify that union layout produces same unique files either way

```python
query1 = [("\soms\s",),("\sosa\s",),("spead",),]
query2 = [("\soms\s","\sosa\s","spead",),]
```

```python
results1 = cobj.make_query(query1)
results2 = cobj.make_query(query2)
```

```python
unique_docs1 = set()
for term in results1[0].keys():
    print("The term", term, "shows up in", len(results1[0][term].keys()), "documents.")
    unique_docs1 = unique_docs1.union(set(list(results1[0][term].keys())))

print(f"\nThe unioned terms show up in {len(unique_docs1)} unique documents.")
unique_docs1
```

```python
unique_docs2 = set()
for key in results2[0].keys():
    print("The term", key, "shows up in", len(results2[0][key].keys()), "documents.")
    unique_docs2 = unique_docs2.union(set(list(results2[0][key].keys())))

print(f"\nThe unioned terms show up in {len(unique_docs2)} unique documents.")
unique_docs2
```

```python
# Verify that the queries are equivalent
unique_docs1 == unique_docs2
```

### Run and verify that case sensitivity is added and working

```python
query = [("OMS",),("OSA",),("SPEAD",),]

results1 = cobj.make_query(query)  # case insensitive
results2 = cobj.make_query(query, case_sensitive=True)  # case sensitive  

print("\nCase Insensitive Results:")
unique_docs1 = set()
for term in results1[0].keys():
    print("The term", term, "shows up in", len(results1[0][term].keys()), "documents.")
    unique_docs1 = unique_docs1.union(set(list(results1[0][term].keys())))

print(f"The unioned terms show up in {len(unique_docs1)} unique documents.\n")
unique_docs1

print("\nCase Sensitive Results:")
unique_docs2 = set()
for key in results2[0].keys():
    print("The term", key, "shows up in", len(results2[0][key].keys()), "documents.")
    unique_docs2 = unique_docs2.union(set(list(results2[0][key].keys())))

print(f"The unioned terms show up in {len(unique_docs2)} unique documents.\n")
unique_docs2
```

```python
def write_results_file(query_results):
    import json

    class SetEncoder(json.JSONEncoder):
        def default(self, obj):
            if isinstance(obj, set):
                return sorted(list(obj))
            return json.JSONEncoder.default(self, obj)

    with open("query_results.json", 'w') as jf:
        json.dump(query_results, jf, indent=4, cls=SetEncoder)


write_results_file(results[0])
```

```python
# remove instances where it shows up on less than 1 page 
results_copy = results[0].copy()
for term in results_copy.keys():
    for document, pages in results_copy[term].copy().items():
        if len(pages) < 2:
            del results_copy[term][document]

significant_unique_docs = set()
for term in results_copy.keys():
    print("The term", term, "shows up in", len(results_copy[term].keys()), "documents.")
    significant_unique_docs = significant_unique_docs.union(set(list(results_copy[term].keys())))

print(f"\nThe unioned terms show up in {len(significant_unique_docs)} unique documents with more than 1 appearance.")
results_copy
```
